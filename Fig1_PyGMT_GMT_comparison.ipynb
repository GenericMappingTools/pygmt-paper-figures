```python
# The PyGMT version:
import pygmt

fig = pygmt.Figure()
fig.basemap(projection="N12c", region=[0, 360, -90, 90], frame=True)
fig.coast(land="tan", water="lightblue")
fig.text(position="MC", text="GMT", font="80p,Helvetica-Bold,red@75")
fig.show()
fig.savefig("PyGMT_map.png")
```

```python
%%bash
# The GMT CLI version:
gmt begin GMT_map png
    gmt basemap -JN12c -R0/360/-90/90 -B
    gmt coast -Gtan -Slightblue
    echo GMT | gmt text -F+f80p,Helvetica-Bold,red@75+cMC
gmt end show
```

```python
# Sequence of strings for the PyGMT code.
pygmt_str = """
import pygmt

fig = pygmt.Figure()
fig.basemap(projection="N12c", region=[0, 360, -90, 90], frame=True)
fig.coast(land="tan", water="lightblue")
fig.text(position="MC", text="GMT", font="80p,Helvetica-Bold,red@@75")
fig.show()
fig.savefig("PyGMT_map.png")
""".splitlines()
```

```python
# Sequence of strings for the GMT CLI code.
# The leading tab characters are intentional to match the indentation.
gmt_str = """
gmt begin GMT_map png
\tgmt basemap -JN12c -R0/360/-90/90 -B
\tgmt coast -Gtan -Slightblue
\techo GMT | gmt text -F+f80p,Helvetica-Bold,red@@75+cMC
gmt end show
""".splitlines()
```

```python
import pygmt

para_args = {
    "x": 0.1,
    "y": 0.1,
    "parwidth": "11.75c",
    "linespacing": 0.35,
    "alignment": "left",
    "justify": "BL",
    "font": "Courier-Bold,8p",
    "pen": "1p",
    "fill": "lightgreen",
}

fig = pygmt.Figure()
fig.basemap(region=[0, 12.5, 0, 2.1], projection="x1c", frame="none")
fig.paragraph(text=gmt_str, **para_args)
fig.text(text="(a) GMT CLI", position="TL", font="9p,1", offset="0/0.3", no_clip=True)

fig.shift_origin(yshift="-h-1.25c")
fig.basemap(region=[0, 12.5, 0, 2.8], projection="x1c", frame="none")
fig.paragraph(text=pygmt_str, **para_args)
fig.text(text="(b) PyGMT", position="TL", font="9p,1", offset="0/0.3", no_clip=True)

fig.shift_origin(yshift="-h-4c")

fig.basemap(projection="N12c", region=[0, 360, -90, 90], frame=True)
fig.coast(land="tan", water="lightblue")
fig.text(position="MC", text="GMT", font="80p,Helvetica-Bold,red@75")
fig.text(text="(c)", position="TL", font="9p,1", offset="0/0.5", no_clip=True)
       
fig.show()
fig.savefig("Fig1_PyGMT_GMT_comparison.png")
```
