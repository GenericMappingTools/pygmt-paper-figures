```python
# The PyGMT version:
import pygmt

fig = pygmt.Figure()
fig.basemap(projection="R7c", region=[0, 360, -90, 90], frame=True)
fig.coast(land="tan", water="lightblue")
fig.text(position="MC", text="PyGMT", font="40p,AvantGarde-Book,red@75")
fig.show()
fig.savefig("PyGMT_map.png")
```

```python
%%bash
# The GMT CLI version:
gmt begin GMT_map png
    gmt basemap -JR7c -R0/360/-90/90 -B
    gmt coast -Gtan -Slightblue
    echo PyGMT | gmt text -F+f40p,AvantGarde-Book,red@75+cMC
gmt end show
```

```python
# Sequence of strings for the PyGMT code.
pygmt_str = """
import pygmt

fig = pygmt.Figure()
fig.basemap(projection="R7c", region=[0, 360, -90, 90], frame=True)
fig.coast(land="tan", water="lightblue")
fig.text(position="MC", text="PyGMT", font="40p,AvantGarde-Book,red@@75")
fig.show()
fig.savefig("PyGMT_map.png")
""".splitlines()
```

```python
# Sequence of strings for the GMT CLI code.
# The leading tab characters are intentional to match the indentation.
gmt_str = """
gmt begin GMT_map png
\tgmt basemap -JR7c -R0/360/-90/90 -B
\tgmt coast -Gtan -Slightblue
\techo PyGMT | gmt text -F+f40p,AvantGarde-Book,red@@75+cMC
gmt end show
""".splitlines()
```

```python
import pygmt

para_args = {
    "parwidth": "12.2c",
    "linespacing": 0.35,
    "alignment": "left",
    "justify": "BL",
    "font": "Courier-Bold,8p",
    "pen": "1p",
    "fill": "bisque",
}

fig = pygmt.Figure()
fig.basemap(region=[-0.1, 12.5, -0.1, 6], projection="x1c", frame="none")
fig.paragraph(x=0, y=3.3, text=gmt_str, **para_args)
fig.text(x=0, y=5.3, text="(a) GMT CLI", justify="BL", font="9p,Helvetica-Bold")

fig.paragraph(x=0, y=0, text=pygmt_str, **para_args)
fig.text(x=0, y=2.75, text="(b) PyGMT", justify="BL", font="9p,Helvetica-Bold")

fig.shift_origin(xshift="w+0.6c", yshift=0.6)

fig.basemap(projection="R7c", region=[0, 360, -90, 90], frame=True)
fig.coast(land="tan", water="lightblue")
fig.text(position="MC", text="PyGMT", font="40p,AvantGarde-Book,red@75")
fig.text(text="(c)", position="TL", font="9p,Helvetica-Bold", offset="-0.5/0.6", no_clip=True)
       
fig.show()
fig.savefig("Fig1_PyGMT_GMT_comparison.png")
```
