```python
import pygmt
from pygmt.params import Position

mars = pygmt.datasets.load_mars_relief(resolution="10m")
with pygmt.config(PROJ_ELLIPSOID="mars"):
    mars_filter = mars.gmt.filter(filter="g50", distance=4)
    mars_grad = mars_filter.gmt.gradient(azimuth=-45, normalize="cauchy")

fig = pygmt.Figure()
pygmt.makecpt(cmap="batlow", series=[-6500, 6500])
fig.grdimage(mars_filter, shading=mars_grad, projection="G-90/20/12c", frame="g30")
fig.shift_origin(xshift="w+0.5c")
fig.grdimage(mars_filter, shading=mars_grad, projection="G90/-20/12c", frame="g30")
fig.colorbar(
    frame=["xa2000f1000+lElevation", "y+lm"],
    position=Position("BC", cstype="outside", offset=(-6.25, 0.2)),
    bg_triangle=True,
    fg_triangle=True,
    move_text="label",
)
fig.show()
# fig.savefig(fname="Fig5_PyGMT_xarray.png")
```
