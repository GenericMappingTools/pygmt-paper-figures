```python
# The PyGMT version:
import pygmt

fig = pygmt.Figure()
fig.basemap(projection="N15c", region=[0, 360, -90, 90], frame=True)
fig.coast(land="tan", water="lightblue")
fig.text(position="MC", text="PyGMT", font="80p,Helvetica-Bold,red@75")
fig.show()
fig.savefig("PyGMT-map.png")
```

```python
%%bash
# The GMT CLI version:
gmt begin GMT-map png
    gmt basemap -JN15c -R0/360/-90/90 -B
    gmt coast -Gtan -Slightblue
    echo GMT | gmt text -F+f80p,Helvetica-Bold,red@75+cMC
gmt end show
```

```python
# Sequence of strings for the PyGMT code.
pygmt_str = """
import pygmt

fig = pygmt.Figure()
fig.basemap(projection="N15c", region=[0, 360, -90, 90], frame=True)
fig.coast(land="tan", water="lightblue")
fig.text(position="MC", text="PyGMT", font="80p,Helvetica-Bold,red@75")
fig.show()
fig.savefig("PyGMT-map.png")
""".splitlines()
```

```python
# Sequence of strings for the GMT CLI code.
# The leading tab characters are intentional to match the indentation.
gmt_str = """
gmt begin GMT-map png
\tgmt basemap -JN15c -R0/360/-90/90 -B
\tgmt coast -Gtan -Slightblue
\techo GMT | gmt text -F+f80p,Helvetica-Bold,red@75+cMC
gmt end show
""".splitlines()
```

```python
import pygmt

fig = pygmt.Figure()
pygmt.config(FONT_TITLE="10p", MAP_TITLE_OFFSET="4p")
fig.basemap(region=[0, 20, 0, 3], projection="X7.25/2c", frame="lrtb+tGMT CLI")
fig.paragraph(
    x=0.25,
    y=0.6,
    text=gmt_str,
    parwidth=50,
    linespacing=0.25,
    alignment="left",
    justify="BL",
    font="Courier-Bold,6p"
)

fig.shift_origin(xshift="w+0.1c")
fig.basemap(region=[0, 20, 0, 3], projection="X9.25/2c", frame="lrtb+tPyGMT")
fig.paragraph(
    x=0.25,
    y=0.25,
    text=pygmt_str,
    parwidth=50,
    linespacing=0.25,
    alignment="left",
    justify="BL",
    font="Courier-Bold,6p"
)
fig.show()
fig.savefig("pygmt-gmt-comparison.png")
```
