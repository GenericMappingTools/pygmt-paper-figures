```python
# Import the PyGMT library
import pygmt

region=[127.5, 128.5, 26, 27]
projection="M10c"


# Use the Figure class to create a figure object, named "fig"
fig = pygmt.Figure()
fig.coast(
    region=region,
    projection=projection,
    # Use default setting for figure's frame
    frame=True,
    # Set the color of the land in the figure
    land="lightgreen",
    # Set the color of the water in the figure
    water="lightblue",
    # Set the thickness of the shorelines
    shorelines="0.5p",
)
# Display the figure object
fig.show()
```
